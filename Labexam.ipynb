```python
from collections import deque

def pour(state, capacities, from_i, to_i):

    state = list(state)
    amount = min(state[from_i], capacities[to_i] - state[to_i])
    state[from_i] -= amount
    state[to_i] += amount
    return tuple(state)

def possible_states(capacities, start):

    visited = set()
    queue = deque([start])
    visited.add(start)

    while queue:
        state = queue.popleft()
        for i in range(3):
            for j in range(3):
                if i != j:
                    new_state = pour(state, capacities, i, j)
                    if new_state not in visited:
                        visited.add(new_state)
                        queue.append(new_state)
    return visited

def main():
    capacities = (12, 8, 5)
    start = (12, 0, 0)
    states = possible_states(capacities, start)

    print("All reachable states (12L, 8L, 5L):")
    for s in sorted(states):
        print(s)
    six_states = [s for s in states if 6 in s]
    print("\nStates where exactly 6 liters is measurable:")
    for s in six_states:
        print(s)


    measurable = set()
    for s in states:
        measurable.update(s)
    measurable = sorted(measurable)

    print("\nAll measurable quantities:", measurable)

if __name__ == "__main__":
    main()
```
